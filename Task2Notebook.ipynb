```python
# Create a function to define those who have lower limit (less than 100K)
def person(client):
    education, sex = client
    # X3: Education (1 = graduate school; 2 = university; 3 = high school; 0, 4, 5, 6 = others).
    if education = 0:
        return '0'
    if education = 1:
        return '1'   
    if education = 2:
        return '2'
    if education = 0:
        return '0'
    if education = 4:
        return 'other'
    if education > 100000 and limit < 300000:
        return 'MEDIUM'
    else:
        return 'HIGH'
    
```

```python
# Create a function to define those who have lower limit (less than 100K)
def lower_limit(client):
    limit, sex = client
    
    if limit < 100000:
        return 'LOW'
    if limit >= 100000 and limit < 300000:
        return 'MEDIUM'
    else:
        return 'HIGH'
```

```python
# Create a function to define those who have lower limit (less than 100K)
def age_range(client):
    age, sex = client
    
    if age < 30:
        return '20'
    if age >= 30 and age < 40:
        return '30'
    if age >= 40 and age < 50:
        return '40'
    if age >= 50 and age < 60:
        return '50'
    if age >= 60 and age < 70:
        return '60'
    if age >= 70 and age < 80:
        return '70'
    else:
        return '80'
```

```python
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import RFE
from sklearn.linear_model import RidgeCV, LassoCV, Ridge, Lasso
```

```python
#Loading the dataset

X = credit.drop("Ydefault_payment_next_month",1)   #Feature Matrix
y = credit["Ydefault_payment_next_month"]          #Target Variable
```

```python
reg = LassoCV()
reg.fit(X, y)
print("Best alpha using built-in LassoCV: %f" % reg.alpha_)
print("Best score using built-in LassoCV: %f" %reg.score(X,y))
coef = pd.Series(reg.coef_, index = X.columns)
```

```python
print("Lasso picked " + str(sum(coef != 0)) + " variables and eliminated the other " +  str(sum(coef == 0)) + " variables")
```

```python
imp_coef = coef.sort_values()
import matplotlib
matplotlib.rcParams['figure.figsize'] = (8.0, 10.0)
imp_coef.plot(kind = "barh")
plt.title("Feature importance using Lasso Model")
```

```python
#Backward Elimination
import statsmodels.api as sm
#Adding constant column of ones, mandatory for sm.OLS model
X_1 = sm.add_constant(X)
#Fitting sm.OLS model
model = sm.OLS(y,X_1).fit()
model.pvalues
```

```python
#Backward Elimination
cols = list(X.columns)
pmax = 1
while (len(cols)>0):
    p= []
    X_1 = X[cols]
    X_1 = sm.add_constant(X_1)
    model = sm.OLS(y,X_1).fit()
    p = pd.Series(model.pvalues.values[1:],index = cols)      
    pmax = max(p)
    feature_with_p_max = p.idxmax()
    if(pmax>0.05):
        cols.remove(feature_with_p_max)
    else:
        break
selected_features_BE = cols
print(selected_features_BE)
```

```python
#RFE (Recursive Feature Elimination)
model = LinearRegression()
#Initializing RFE model
rfe = RFE(model, 7)
#Transforming data using RFE
X_rfe = rfe.fit_transform(X,y)  
#Fitting the data to model
model.fit(X_rfe,y)
print(rfe.support_)
print(rfe.ranking_)
```

```python
#RFE (Recursive Feature Elimination)
#no of features
nof_list=np.arange(1,23)            
high_score=0
#Variable to store the optimum features
nof=0           
score_list =[]
for n in range(len(nof_list)):
    X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state = 0)
    model = LinearRegression()
    rfe = RFE(model,nof_list[n])
    X_train_rfe = rfe.fit_transform(X_train,y_train)
    X_test_rfe = rfe.transform(X_test)
    model.fit(X_train_rfe,y_train)
    score = model.score(X_test_rfe,y_test)
    score_list.append(score)
    if(score>high_score):
        high_score = score
        nof = nof_list[n]
print("Optimum number of features: %d" %nof)
print("Score with %d features: %f" % (nof, high_score))
```
